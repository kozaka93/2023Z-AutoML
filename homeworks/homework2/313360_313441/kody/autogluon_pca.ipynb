```python
import pandas as pd
df_ultimate_test = pd.read_table("artificial_test.data", sep=" ", header=None)
df_ultimate_test = df_ultimate_test.drop(df_ultimate_test.columns[500], axis=1) # kolumna 500 do wyrzucenia
df_ultimate_test
```

```python
df_data_X = pd.read_table("artificial_train.data", sep=" ", header=None)
df_data_X = df_data_X.drop(df_data_X.columns[500], axis=1)
df_data_X
```

```python
df_data_Y = pd.read_table("artificial_train.labels", sep=" ", header=None)
df_data_Y.columns = ["label"]
df_data_Y
```

```python
data_concat = pd.concat([df_data_X, df_data_Y], axis=1)
data_concat
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(n_components=7, svd_solver='full')
pca2 = PCA(n_components=20, svd_solver='full')
pca3 = PCA(n_components=50, svd_solver='full')
df_data_X_transformed_1 = pd.DataFrame(pca1.fit_transform(df_data_X, df_data_Y))
df_data_X_transformed_2 = pd.DataFrame(pca2.fit_transform(df_data_X, df_data_Y))
df_data_X_transformed_3 = pd.DataFrame(pca3.fit_transform(df_data_X, df_data_Y))
df_data_X_transformed_1
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data_X, df_data_Y, test_size=0.33, random_state=4, stratify=df_data_Y)
X_train1, X_test1, y_train1, y_test1 = train_test_split(df_data_X_transformed_1, df_data_Y, test_size=0.33, random_state=4, stratify=df_data_Y)
X_train2, X_test2, y_train2, y_test2 = train_test_split(df_data_X_transformed_2, df_data_Y, test_size=0.33, random_state=4, stratify=df_data_Y)
X_train3, X_test3, y_train3, y_test3 = train_test_split(df_data_X_transformed_3, df_data_Y, test_size=0.33, random_state=4, stratify=df_data_Y)
X_train1
```

```python
data_concat_test = pd.concat([X_test, y_test], axis=1)
data_concat_train = pd.concat([X_train, y_train], axis=1)

data_concat_test1 = pd.concat([X_test1, y_test1], axis=1)
data_concat_train1 = pd.concat([X_train1, y_train1], axis=1)

data_concat_test2 = pd.concat([X_test2, y_test2], axis=1)
data_concat_train2 = pd.concat([X_train2, y_train2], axis=1)

data_concat_test3 = pd.concat([X_test3, y_test3], axis=1)
data_concat_train3 = pd.concat([X_train3, y_train3], axis=1)

data_concat_test1
```

```python
label = 'label'
```

```python
predictor1 = TabularPredictor(label=label).fit(data_concat_train1, time_limit=1200, auto_stack=True)
```

```python
predictor1.evaluate(data_concat_test1, silent=True)
```

```python
predictor1.leaderboard(data_concat_test1)
```

```python
predictor1.info()
```
