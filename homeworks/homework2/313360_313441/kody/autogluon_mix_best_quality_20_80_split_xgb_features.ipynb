```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from xgboost import XGBRegressor

class XGBFeatureExtractor(BaseEstimator, TransformerMixin):
    def __init__(self, n_components=100, xgb_params=None):
        """
        XGBFeatureExtractor constructor.

        Parameters:
        - n_components: Number of components to keep.
        - xgb_params: Parameters for XGBoost model.
        """
        self.n_components = n_components
        self.xgb_params = xgb_params
        self.feature_importances_ = None
        self.xgb_model = None

    def fit(self, X, y=None):
        """
        Fit the XGBFeatureExtractor on the input data.

        Parameters:
        - X: Input data.
        - y: Target variable.

        Returns:
        - self
        """
        self.xgb_model = XGBRegressor(**self.xgb_params)
        self.xgb_model.fit(X, y)
        self.feature_importances_ = self.xgb_model.feature_importances_

        return self

    def transform(self, X):
        """
        Transform the input data using the fitted XGBFeatureExtractor.

        Parameters:
        - X: Input data.

        Returns:
        - Transformed data.
        """
        feature_importances = self.feature_importances_
        sorted_indices = np.argsort(feature_importances)[::-1][:self.n_components]
        selected_features = X.iloc[:, sorted_indices]

        return selected_features
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
```

```python
extractor = XGBFeatureExtractor(n_components=20, xgb_params={'learning_rate': 0.1, 'max_depth': 3,
                                 'n_estimators': 100})
df_data_X_transformed_0 = pd.DataFrame(extractor.fit_transform(df_data_X, df_data_Y))
df_data_X_transformed_0
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data_X_transformed_0, df_data_Y, test_size=0.2, random_state=4, stratify=df_data_Y)
```

```python
data_concat_test = pd.concat([X_test, y_test], axis=1)
data_concat_train = pd.concat([X_train, y_train], axis=1)


data_concat_test
```

```python
label = 'label'
```

```python
predictor = TabularPredictor(label=label, eval_metric='balanced_accuracy').fit(data_concat_train, presets=['best_quality'], time_limit=14400)
```

```python
predictor.evaluate(data_concat_test, silent=True)
```
